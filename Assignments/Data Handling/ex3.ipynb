# 7. Select 100th row and all column using iloc function

```python
titanic.iloc[100,:]
```

# 8. Select 100th row and all column using loc function

```python
titanic.loc[100,:]
```

# 9. Select all row with label name and fare column using iloc 

```python
titanic.iloc[:,[2,8]]
```

# 10. Select all row with label name and fare column using loc 

```python
titanic.loc[:,['name','fare']]
```

# 11. Select particular row and column using loc and iloc

```python
titanic.loc[[50,25,15],['pclass','fare','age']]
```

```python
titanic.iloc[[50,25,15],[0,8,4]]
```

# 12. Select particular row and column using loc and iloc

```python
titanic.loc[10:25,['pclass','fare','age']]
```

```python
titanic.iloc[10:26,[0,8,4]]
```

# 13. Select particular row and column using loc and iloc

```python
titanic.loc[10:15,'pclass':'age']
```

```python
titanic.iloc[10:16,0:5]
```

# 14. titanic age above 35

```python
titanic[titanic['age']>=35]
```

# 15. Age >=35 and pclass,age

```python
titanic.loc[(titanic['age']>=35),'pclass':'age']
```

# 16. Select female above 35 years

```python
titanic.loc[(titanic['age']>=35) & (titanic['sex']=='female')]
```
